```python
from pydaqmx_helper.counter import Counter
from pydaqmx_helper.adc import ADC

counter = Counter()
adc = ADC()
adc.addChannels([1], minRange = -1, maxRange = 1)
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf
from scipy.stats import gaussian_kde
import scipy.stats as stats
```

```python
def take_counts(iterations, save_data = False, save_name = "counts.txt"):

    count_array = np.array([])

    for i in np.arange(0, iterations):

        counter.start()
        time.sleep(0.1)
        count = counter.stop()

        count_array = np.append(count_array, count)

    if save_data == True:
        np.savetxt(save_name,
                  np.column_stack([count_array]))
        
def take_time_intervals(num_points, save_data = False, save_name = "time_intervals.txt"):

    interval_array = []
    
    counter.stop()
    counter.start()
    
    old_t = time.perf_counter()
    count = counter.getCount()
    while len(interval_array) < num_points:

        if new_count := counter.getCount() > count:
            
            t = time.perf_counter()
            dt = t - old_t
            old_t = t
            
            count = counter.getCount()
            
            interval_array.append(dt)
        

    if save_data == True:
        np.savetxt(save_name,
                  np.column_stack([interval_array]))
        
    return interval_array[1:]

def take_voltage_readings(n, save_data = False, save_name = "voltage_readings.txt"):

    data = adc.sampleVoltages(n, 10000)
    data1 = data[1]

    if save_data == True:
        np.savetxt(save_name,
                  np.column_stack([data1]))
        
    return data1
```

```python
no_filter = np.loadtxt("no_filter_voltage_readings.txt")
data_files = ["1kohms1microFarad_voltage_readings.txt",
             "150ohms1microFarad_voltage_readings.txt",
             "150ohms10microFarad_voltage_readings.txt",
             "5500ohms0point1microFarad_voltage_readings.txt"]

RC1 = 1000 * 10e-6
RC2 = 150 * 10e-6
RC3 = 150 * 10e-5
RC4 = 5500 * (0.1*10e-6)

print(RC1, RC2, RC3, RC4)

fig = plt.figure(constrained_layout = True, figsize = (12, 6))
axd = fig.subplot_mosaic(
            """
            ABC
            ADE
            """
)

keys = ["A", "B", "C", "D", "E"]

# Plot autocorrelation function with no filter.
plot_acf(np.asarray(no_filter),
         title = "",
         ax = axd["A"],
         fft = True,
         alpha = None,
         vlines_kwargs = {"color": "black"},
         **{"color": "black", "markersize": 4})

# Plot autocorrelation function with filters.
for file, key in zip(data_files, keys[1:]):
    
    data = np.loadtxt(file)
    
    plot_acf(np.asarray(data),
         title = "",
         ax = axd[key],
         fft = True,
         alpha = None,
         vlines_kwargs = {"color": "black"},
         **{"color": "black", "markersize": 4})

for key in keys:
    axd[key].grid(linestyle = "--")
    axd[key].set_ylim(-0.1, 1.1)
    
fig.supxlabel("Time")
fig.supylabel("Autocorrelation function")
    
plt.savefig("Autocorrelation.pdf", bbox_inches = "tight")
```
